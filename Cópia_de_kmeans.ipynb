```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
curva_abc = pd.read_csv('https://raw.githubusercontent.com/sanchesprogramer/Ciencia-de-dados/main/trab11.csv', encoding='UTF-8', sep=';')
curva_abc.head()
curva_abc.shape

X = curva_abc['Custo (R$)']
y = curva_abc['Movimentação']

plt.scatter(X, y)
```

```python
curva_abc = curva_abc[['Custo (R$)', 'Movimentação']]
kmeans = KMeans(n_clusters=3, random_state=0).fit(curva_abc)
```

```python
#clustering_model.labels_
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
data_labels = kmeans.labels_
X = curva_abc['Custo (R$)']
y = curva_abc['Movimentação']
plt.scatter(X, y, c = data_labels, label=data_labels)
```
